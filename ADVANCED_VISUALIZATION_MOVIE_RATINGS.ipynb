```python
# How to work with joint plots
from matplotlib import pyplot as plt # visualization
import seaborn as sns # advanced visualization

import warnings
warnings.filterwarnings('ignore')
```

```python
j=sns.jointplot(data=movies,x='CriticRating',y='AudienceRating') 

# Audience rating is more dominant then critics rating
# Base on this we find out as most people are most liklihood to watch audience rating & less likely to watch critics rating
# there is postive corelation between 2 attributes
```

```python
j=sns.jointplot(data=movies,x='CriticRating',y='AudienceRating',kind='hex') 
```
