```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import preprocessing
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.metrics import r2_score
```

```python
data=pd.read_csv(r"C:\Users\divya\Downloads\car-mpg.csv")
data.head()
```

```python
# drop car name

data=data.drop(['car_name'],axis=1)
data['origin']=data['origin'].replace({1:'america',2:'europe',3:'asia'})
data=pd.get_dummies(data,columns=['origin'],dtype=int)
```

```python
data.head()
```

```python
data=data.replace('?',np.nan)
```

```python
data = data.apply(lambda x: pd.to_numeric(x, errors='coerce').fillna(pd.to_numeric(x, errors='coerce').median()) if x.dtype == object else x.fillna(x.median()), axis=0)
```

```python
data.head()
```

```python
data
```

```python
X=data.drop(['mpg'],axis=1)
y=data[['mpg']]
```

```python
X_s=preprocessing.scale(X)
X_s=pd.DataFrame(X_s,columns=X.columns)

y_s=preprocessing.scale(y)
y_s=pd.DataFrame(y_s,columns=y.columns)
```

```python
X_s
```

```python
y_s
```

```python
  X_train,X_test,y_train,y_test=train_test_split(X_s,y_s,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

# Simple Linear regression 

```python
regression_model=LinearRegression()
regression_model.fit(X_train,y_train)

for idx,col_name in enumerate(X_train.columns):
    print(f'The coefficient for {col_name} is {regression_model.coef_[0][idx]}')

intercept=regression_model.intercept_[0]
print('The intercept is {}'.format(intercept))
```

# Regularized Ridge Regression 

```python
ridge_model=Ridge(alpha=0.4)
ridge_model.fit(X_train,y_train)

print('Ridge model coef: {}'.format(ridge_model.coef_))
```

# Regularized Lasso Regression

```python
lasso_model=Lasso(alpha=0.1)
lasso_model.fit(X_train,y_train)

print('Ridge model coef: {}'.format(lasso_model.coef_))
```
