# SEABORN

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns # distribution visualization
```

```python
%matplotlib inline
plt.rcParams['figure.figsize']=9,5

import warnings
warnings.filterwarnings('ignore') 
```

```python
store.head()
```

```python
store['InternetUsers']
```

```python
vis1=sns.distplot(store['InternetUsers']) # UNI-VARIATE ANALYSIS
plt.show()
```

```python
vis1=sns.distplot(store['InternetUsers'],bins=10)
plt.show()
```

```python
vis2=sns.boxplot(data=store,x="IncomeGroup",y="BirthRate") # BI-VARIATE ANALYSIS
plt.show()
```

```python
vis3=sns.lmplot(data=store,x="InternetUsers",y="BirthRate",fit_reg=True) # BI-VARIATE ANALYSIS
plt.show()
```

```python
vis3=sns.lmplot(data=store,x="InternetUsers",y="BirthRate",fit_reg=False) # BI-VARIATE ANALYSIS
plt.show()
```

```python
vis3=sns.lmplot(data=store,x="InternetUsers",y="BirthRate",fit_reg=False,hue='IncomeGroup') # BI-VARIATE ANALYSIS
plt.show()
```
