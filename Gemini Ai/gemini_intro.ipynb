```python
import textwrap
from IPython.display import display, Markdown

def to_markdown(text):
    # Replace periods with " *"
    text = text.replace('•', ' *')
    # Indent each line with "> "
    indented = textwrap.indent(text, prefix='> ', predicate=lambda _: True)
    return Markdown(indented)

display(to_markdown(response.text))
```
